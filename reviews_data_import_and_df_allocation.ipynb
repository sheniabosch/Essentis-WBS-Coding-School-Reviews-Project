# Clean Reviews DF

## Import Clean Reviews CSV

```python
with open('/content/drive/MyDrive/Essentis/data/reviews_clean.csv', 'r') as f:
    for i in range(10):
        print(f"{i+1}: {f.readline()}")
```

## Create Clean Reviews Dataframe

```python
# specify delimeriters, since our source data is formatted inconsistently
import pandas as pd

clean_reviews_df = pd.read_csv(
    '/content/drive/MyDrive/Essentis/data/reviews_clean.csv',
    sep=';',                # semicolon delimiter
    engine='python',        # better handling of multi-line text
    quotechar='"',          # handle quotes properly
    skiprows=1,             # skip first row ("Table 1")
    on_bad_lines='warn'     # warns about bad lines but continues
)
```

sep=';' → semicolon-delimited

engine='python' → more forgiving parser

quotechar='"' → handle quotes properly

on_bad_lines='skip' → optional, skips truly broken rows

skiprows=1 → if you want to ignore that first line "Table 1"

```python
print(clean_reviews_df.shape)
clean_reviews_df.head()
```

# New Reviews DF

## Import New Reviews CSV

```python
with open('/content/drive/MyDrive/Essentis/data/reviews_new.csv', 'r') as f:
    for i in range(10):
        print(f"{i+1}: {f.readline()}")
```

## Create New Reviews Dataframe

```python
import pandas as pd

new_reviews_df = pd.read_csv(
    '/content/drive/MyDrive/Essentis/data/reviews_new.csv',
    sep=';',                     # semicolon delimiter
    engine='python',             # handle multi-line text safely
    quotechar='"',               # handle quotes properly
    skiprows=1,                  # skip first row ("Table 1")
    on_bad_lines='warn'          # warn about broken lines but continue
)

# Optional: drop empty trailing columns
new_reviews_df = new_reviews_df.loc[:, new_reviews_df.columns.str.strip() != '']

# Quick check
print(new_reviews_df.shape)
print(new_reviews_df.columns)
new_reviews_df.head()
```

engine='python' allows pandas to handle line breaks inside quotes, which the default C engine chokes on.

sep=';' fixes the field parsing.

skiprows=1 ignores the "Table 1" line.

on_bad_lines='warn' ensures it won’t crash if a tiny row is malformed.

Dropping empty columns fixes trailing semicolons in the header.

the reviews_new.csv is now new_reviews_df

# Combined DF

```python
# adds a source column to the master dataframe
google_reviews_df['data_source'] = 'google'
clean_reviews_df['data_source'] = 'clean'
new_reviews_df['data_source'] = 'new'
```

```python
import pandas as pd

master_reviews_df = pd.concat(
    [google_reviews_df, clean_reviews_df, new_reviews_df],
    ignore_index=True,   # resets the row index
    sort=False           # keeps columns as-is if some columns differ
)
```

```python
print(master_reviews_df.shape)       # total rows and columns
print(master_reviews_df.columns)     # check all column names
master_reviews_df['source'].value_counts()  # see how many rows came from each CSV
```

This shows all the columns in all of our datasets- giving an overview of inconsistencies which need cleaning.

## Create Master Dataframe CSV

```python
master_reviews_df.to_csv(
    '/content/drive/MyDrive/Essentis/data/master_reviews.csv',
    index=False,
    sep=';',  # keep semicolon consistent if needed
    quoting=1  # quote all fields to be safe
)
```

This master dataframe can now be read as a csv and used for cleaning the data.
